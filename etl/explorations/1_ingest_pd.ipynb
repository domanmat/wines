# Cleaning and casting

## Casting column types

```python
cast_columns = [
    ("alcohol",         "float"),
    ("vintage",         "int"),
    ("case_production", "int"),
    ("retail",          "float"),
    ("rating",          "int"),
    ("bottle_size",     "float"),
]

df_cast = df.copy()

for column, dtype in cast_columns:
    nulls_before = df_cast[column].isna().sum()
    cast_col = pd.to_numeric(df_cast[column], errors="coerce")
    nulls_after = cast_col.isna().sum()
    error_mask = df_cast[column].notna() & cast_col.isna()
    failures = error_mask.sum()

    df_cast[f"{column}_cast"] = cast_col
    df_cast[f"{column}_error"] = df_cast[column].where(error_mask)

    print(f"{column:20s}  nulls before={nulls_before:>6,}  after={nulls_after:>6,}  failures={failures:>6,}")
    if failures > 0:
        print(df_cast.loc[error_mask, column].unique())
```

## Special cleaning and feature creation

### alcohol

```python
df_cast["alcohol_cast"].value_counts(dropna=False).head(1000)
```

```python
df_cast["alcohol_cast"].value_counts(dropna=False).sort_index(ascending=False).head(1000)
```

```python
def fix_alcohol(v):
    if pd.isna(v):
        return np.nan
    if v > 25.0:
        return np.nan
    if v > 7.0:
        return v / 100
    if v > 2.5:
        return np.nan
    if v > 0.7:
        return v / 10
    if v > 0.25:
        return np.nan
    if v >= 0.05:
        return v
    return np.nan

df_cast["alcohol"] = df_cast["alcohol_cast"].map(fix_alcohol)
```

### bottle_size

```python
df_cast["bottle_size"].value_counts(dropna=False).head(1000)
```

```python
def parse_bottle_size(s):
    if pd.isna(s):
        return np.nan
    s = str(s).strip()
    s_lower = s.lower()
    if s_lower.endswith("l ml"):
        try:
            return float(s[:-4].strip()) * 1000
        except ValueError:
            return np.nan
    elif s_lower.endswith("ml"):
        try:
            return float(s[:-2].strip())
        except ValueError:
            return np.nan
    elif s_lower.endswith("l"):
        try:
            return float(s[:-1].strip()) * 1000
        except ValueError:
            return np.nan
    else:
        try:
            return float(s)
        except ValueError:
            return np.nan

df_cast["bottle_size"] = df_cast["bottle_size"].map(parse_bottle_size)
```

```python
valid_sizes = {750.0, 375.0, 500.0, 1500.0, 3000.0, 250.0, 1000.0, 187.0, 620.0}

def clean_bottle_size(v):
    if pd.isna(v):
        return np.nan
    if v == 75.0:
        return 750.0
    if v == 751.0:
        return 750.0
    if v == 740.0:
        return 750.0
    if v in valid_sizes:
        return v
    return np.nan

df_cast["bottle_size"] = df_cast["bottle_size"].map(clean_bottle_size)
```
