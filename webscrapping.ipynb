```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.wineenthusiast.com/buying-guide/tussock-jumper-2025-sauvignon-blanc-marlborough/'
response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
soup = BeautifulSoup(response.text, 'html.parser')

# Paste the CSS selector from DevTools directly
region_el = soup.select_one(
    "#single-page > header > div > div > div > div > "
    "div.winery-location.d-flex.align-content-center > span:nth-child(2) > a:nth-child(1)"
)

region = region_el.text.strip() if region_el else None
print(region)  # → 'Marlborough'
```

```python
from playwright.async_api import async_playwright
from playwright_stealth import Stealth
from bs4 import BeautifulSoup

url = 'https://www.wineenthusiast.com/buying-guide/tussock-jumper-2025-sauvignon-blanc-marlborough/'

async def scrape():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=False)
        async with await Stealth().use_async(browser.new_context()) as context:
            page = await context.new_page()
            await page.goto(url, wait_until='networkidle')
            html = await page.content()
        await browser.close()
    return html

html = await scrape()

soup = BeautifulSoup(html, 'html.parser')

region_el = soup.select_one(
    "#single-page > header > div > div > div > div > "
    "div.winery-location.d-flex.align-content-center > span:nth-child(2) > a:nth-child(1)"
)

region = region_el.text.strip() if region_el else None
print(region)
```
